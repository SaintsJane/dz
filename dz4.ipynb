```python
import numpy as np
A = np.array([[1, 2,1], [-2,1, 1]])
print ("Матрица A:\n", A)
B = np.array([[3,1], [-1,0], [2,2]])
print ("Матрица B:\n", B)
C = np.array([[-2,2], [1,-1], [1,2]])
print ("Матрица A транспонированная:\n", A)
print ("Сложение разноразмерных матриц не возможно, поэтому транспонируем матрицу А")
At = np.transpose(A)
print ("Матрица A транспонированная:\n", At)
D = At + 2*B
print ("Результат сложения:\n", D)
```

```python
import numpy as np
print("Матрица заполненная значениями в шахматном порядке:\n")
E = np.zeros((8,8),dtype=int)
E[1::2,::2] = 1
E[::2,1::2] = 1
print(E)
```
